# Removal of Punctuation

```python
import string
string.punctuation
```

```python
def remove_punc(text):
  punc = string.punctuation
  return text.translate(str.maketrans("","",punc))
```

```python
data['clean text'] = data['tweet'].apply(lambda x: remove_punc(x))
```

```python
data.head()
```

# convert to lovercase

```python
data['clean text'] = data['tweet'].str.lower()
```

# Removal of Stopwords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#to display all stopwords
",".join(stopwords.words("english"))
```

```python
stopwords = set(stopwords.words("english"))
```

```python
def remove_stopwords(text):
  return "".join([word for word in text.split() if word not in stopwords])
```

```python
data['clean text'] = data['tweet'].apply(lambda x:remove_stopwords(x))
```

```python
data.sample(5)
```

# Removal of Frequent words

```python
from collections import Counter
word_count = Counter()
for text in data['tweet']:
  for word in text.split():
    word_count[word]+=1
```

```python
word_count.most_common(5)
```

```python
freq_word = set(word for (word,wc) in word_count.most_common(5))
```

```python
def remove_freq_words(text):
  return "".join([word for word in text.split() if word not in freq_word])
```

```python
data['clean text'] = data['tweet'].apply(lambda x:remove_freq_words(x))
```

```python
data.sample(5)
```

# Removal of Rare Words

```python
rare_word = set(word for (word,wc)in word_count.most_common()[0:50])
rare_word
```

```python
def remove_rare_word(text):
  return "".join([word for word in text.split () if word not in rare_word])
```

```python
data['clean text'] = data['tweet'].apply(lambda x: remove_rare_word(x))
```

```python
data.head()
```

# Removal of Special characters

```python
import re
```

```python
def remove_spl_chars(text):
  text = re.sub('[^a-zA-Z0-9]',' ',text)
  text = re.sub('\s+',' ',text)
  return text
```

```python
data['clean text'] = data['tweet'].apply(lambda x: remove_spl_chars(x))
```

```python
data.head()
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
data['stemmed_text'] = data['clean text'].apply(lambda x: stem_words(x))
```

```python
data.sample(5)
```

# Removal of URLs

```python
text = """Sustainable tourism has become an increasingly important topic as the impact of travel on the
environment and local communities gains more attention. According to the World Tourism Organization,
sustainable tourism aims to minimize negative effects and enhance the positive contributions of tourism.
For more information on sustainable travel practices, visit sites like https://www.sustainabletravel.org
and https://www.ecotourism.org. Additionally, https://www.greenmatters.com and
https://www.nationalgeographic.com offer insights on how travelers can reduce their carbon footprint.
Governments and organizations worldwide are implementing policies to promote responsible tourism;
details can be found at https://www.unwto.org and https://www.responsibletravel.com.
To learn about initiatives and success stories, websites such as https://www.travelandleisure.com and
https://www.lonelyplanet.com provide excellent resources."""
```

```python
def remove_urls(text):
  return re.sub(r"https?://\S+|www\.\S+",'',text)
```

```python
remove = remove_urls(text)
print(remove)
```

# Removal of HTML tags

```python
text1 = """<p>Text preprocessing is an essential step in natural language processing (NLP) pipelines.
 <strong>Tokenization</strong> breaks text into individual words or tokens, while <em>lemmatization</em>
  reduces words to their base or dictionary form.</p>
<ul>
  <li><a href="https://en.wikipedia.org/wiki/Stop_words" target="_blank">Stop words</a> removal eliminates common
   words like "the," "is," and "and," which often carry little meaning.</li>
  <li><code>Normalization</code> involves converting text to lowercase and removing punctuation to standardize the text data.</li>
</ul>
<div class="note">
  <p>HTML tags can contain valuable information, but they need to be handled appropriately during text preprocessing
   to avoid interfering with analysis.</p>
</div>
"""
```

```python
def remove_html_tags(text):
  return re.sub(r"<.*?>",'',text)
```

```python
RHT = remove_html_tags(text1)
print(RHT)
```
